```python
import math
def count_plots_services(dataframe):
    num_cols = 14
    num_rows = math.ceil(num_cols / 2)  

    fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(16, 4 * num_rows))
    axes = axes.flatten()
    categorical_columns = dataframe.select_dtypes(include=['object', 'category']).columns
    for i, column in enumerate(categorical_columns):
        sns.countplot(data=dataframe, x=column, order=dataframe[column].value_counts().index, palette='bone_r', ax=axes[i])
        axes[i].set_title(f'Count Plot for {column}')
    plt.tight_layout()
    plt.show()
```

```python
count_plots_services(satisfaction_service)
```
